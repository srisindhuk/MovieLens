```python
# 8. Visualize how popularity of genres has changed over the years

releaseyeardf= moviedf.drop(['movie id','movie title'], axis = 1)
releaseyeardf['release date'] = releaseyeardf['release date'].dt.year
releaseyeardf= releaseyeardf.rename({'release date':'release year'}, axis='columns')
releaseyeardf
grouped = releaseyeardf.groupby('release year').sum()
#grouped
grouped['total']=grouped.sum(axis=1)
grouped
```

```python
percentdf = (grouped.divide(grouped['total'],axis=0))*100
```

```python
grouped.drop('total',axis=1)
percentdf
```

```python
sns.heatmap(percentdf, cmap="YlGnBu",linewidths=.3)
```

**insights**
Converted release date into datetime and taken year from it to groupby to get number of movies released per  genre ina year and then got total number of movies released in a year. And finally calculated the %age of each genre. and used a heatmap.


```python
#9.Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times 

merged_df= pd.merge(moviedf,datadf) # merging the two csvs to get movie title and rating in the same df.
movie_ratings = merged_df.groupby('movie title')['rating'].aggregate(['count','mean']) #group by title and then get count of ratings and avg of ratings
hundredrating =movie_ratings['count']>=100 
#hundredrating
movie_ratings[hundredrating].sort_values(by=['mean'], ascending=False)[:25]
```

**insights**
After merging the two dataframes, grouped by 'movie title' an got the count and mean of the rating so that number of ratings given to each movie and average rating for each movie is retrieved.
Then sliced the movies with more than 100 ratings.
Sorted according to mean value to get top avg rated movies. And finally top 25 movies have been printed.
